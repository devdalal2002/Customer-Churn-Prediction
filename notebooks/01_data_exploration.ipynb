# 01 - Data exploration

This notebook performs EDA for the Telco Customer Churn dataset and produces interactive Plotly charts for churn breakdowns.

Make sure you have run `churn_pipeline.py` or created `data/interim/cleaned_churn.csv` before running heavy cells.

```python
# Cell 1: Imports and helper functions
import os
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# profile tool: ydata_profiling (formerly pandas-profiling)
try:
    from ydata_profiling import ProfileReport
except Exception:
    ProfileReport = None

DATA_INTERIM = os.path.join('..','data','interim','cleaned_churn.csv')
DATA_RAW = os.path.join('..','data','raw','Telco-Customer-Churn.csv')


def load_data():
    # prefer interim cleaned if present
    if os.path.exists(DATA_INTERIM):
        print('Loading interim cleaned:', DATA_INTERIM)
        return pd.read_csv(DATA_INTERIM)
    if os.path.exists(DATA_RAW):
        print('Loading raw:', DATA_RAW)
        return pd.read_csv(DATA_RAW)
    raise FileNotFoundError('No data file found. Run churn_pipeline.py to produce interim file or add raw CSV to data/raw.')

print('Imports ready')
```

```python
# Cell 2: Load and initial checks
df = load_data()
print('shape:', df.shape)
print('duplicates:', df.duplicated().sum())
print('missing per column:')
print(df.isna().sum())

# separate categorical vs numerical
num_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols = [c for c in df.select_dtypes(include=['object','category']).columns if c.lower()!='customerid']
print('numerical:', num_cols)
print('categorical:', cat_cols)
```

```python
# Cell 3: Generate profiling report (ydata-profiling)
if ProfileReport is None:
    print('ydata_profiling not available. Install with: pip install ydata-profiling')
else:
    profile = ProfileReport(df, title='Telco Churn Profile', minimal=True)
    out_path = os.path.join('..','reports','figures','telco_profile.html')
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    profile.to_file(out_path)
    print('Saved profile report to', out_path)
```

```python
# Cell 4: Plotly - churn by gender
fig = px.histogram(df, x='gender', color='Churn', barmode='group', title='Churn by Gender',
                   category_orders={'Churn':['No','Yes']})
fig.show()
```

```python
# Cell 5: Plotly - churn by Contract
fig = px.histogram(df, x='Contract', color='Churn', barmode='group', title='Churn by Contract Type')
fig.show()
```

```python
# Cell 6: Plotly - churn by PaymentMethod
fig = px.histogram(df, x='PaymentMethod', color='Churn', barmode='group', title='Churn by Payment Method')
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
# Cell 7: Plotly - churn by InternetService
fig = px.histogram(df, x='InternetService', color='Churn', barmode='group', title='Churn by Internet Service')
fig.show()
```
